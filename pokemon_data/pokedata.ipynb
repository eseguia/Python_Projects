```python
import pandas as pd
from matplotlib import pyplot as plt
import re
```

```python
df = pd.read_csv('pokemon_data.csv')
```

```python
df = df.loc[~df['Name'].str.contains('Mega')]
df
```

```python
def filter_type(t1,t2):
    if t2 == 'any':
        return df.loc[(df['Type 1'] == str(t1))]
    else:
        return df.loc[(df['Type 1'] == str(t1)) & (df['Type 2'] == str(t2))]
    
inp1 = input('Choose a Pokemon Type: \n')
inp2 = input('Choose a pokemon Subtype or "any": \n')
filter_type(inp1,inp2)
```

```python
inp_name = input('Enter a Pokemon name: \n')
df.loc[df['Name'].str.contains(inp_name, flags = re.I, regex = True)]
```

## Download all Pokemon Images and Save them to cd

```python
import requests
import shutil

index = int('000')
for i in range(721):
    index +=1
    ind = ('%03d' % index)
    url = f'https://assets.pokemon.com/assets/cms2/img/pokedex/full/{ind}.png'
    response = requests.get(url)
    with open(f'{ind}.png','wb') as file:
        file.write(response.content)
        file.close()
```

### Show image based on pokemon Name / Index
