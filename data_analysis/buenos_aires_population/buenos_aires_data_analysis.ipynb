```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('censo_PHV_2010.csv')
```

```python
for i in data.DEPTO[0:2480]:
    cant = data[data.DEPTO == i]
    radio = data[data.DEPTO == i]
    rad = radio.RADIO
    pers = cant.PERSONAS
    plt.plot(pers,rad/10**6, 'o')
plt.show()
```

```python
for i in data.DEPTO[0:2480]:
    cant = data[data.DEPTO == i]
    radio = data[data.DEPTO == i]
    rad = radio.RADIO
    hog = cant.HOGARES
    plt.plot(hog,rad/10**6, 'o')
plt.show()
```

```python
data2 = pd.read_csv('buenosaires_population.csv')
plt.plot(data2.Year,data2.Population/10**6)
plt.title('Population of Buenos Aires')
plt.ylabel('Millions of People')
plt.xlabel('Years')
plt.show()
```

```python
data2 = pd.read_csv('buenosaires_population.csv')
plt.plot(data2.Year,data2.Growth/10**6)
plt.title('Growth of population in Buenos Aires')
plt.ylabel('Millions of People')
plt.xlabel('Years')
plt.show()
```

```python
data2 = pd.read_csv('buenosaires_population.csv')
plt.plot(data2.Year,data2.GrowthRate*100)
plt.title('Growth Percentage of population in Buenos Aires')
plt.ylabel('Percentage of growth')
plt.xlabel('Years')
plt.show()
```
