```python
import requests
from bs4 import BeautifulSoup as bs
from matplotlib import pyplot as plt

username = input('Enter GitHub username: \n')
url = f'https://github.com/{username}'
response  = requests.get(url)
soup = bs(response.content)
table = soup.find(class_='js-yearly-contributions')
contributions = table.find(class_='f4 text-normal mb-2').get_text()
day = table.find_all(class_= 'day')
```

```python
day = table.find_all(class_='day')
count = []
for i in day:
    count.append(i['data-count'])
ints = [int(items) for items in count]
plt.ylabel('Contributions')
plt.xlabel('Days')
plt.title(f'Yearly contributions by {username}')
plt.plot(ints)
plt.show()
```
